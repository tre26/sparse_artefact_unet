```python
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
results_folder = "result/"
names = ["artifactual","GT","predicted"]
results = [results_folder + r+ ".mat" for r in names]
```

```python
fig = plt.figure(figsize=(20,20))
axes = fig.subplots(1,3)

for r,n,ax in zip(results,names,axes):
    a = sio.loadmat(r)
    ax.imshow(a[n][0,:,:],cmap="gray",clim=(-0.05,0.05))
    ax.set_title(n)
    ax.axis('off')
plt.show()
```

```python
fig = plt.figure()
axes = fig.subplots(1,3)

for r,n,ax in zip(results,names,axes):
    a = sio.loadmat(r)
    ax.hist(a[n][5,:,:].flatten())
    ax.set_title(n)
plt.show()
```
